```python
# question - 2
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Name': ['Rega', 'Linda', 'John', 'Vishal', 'Mohan'],
    'Age': [40, 26, 21, 23, 35],
    'Salary': [30000, 40000, 100000, 75000, 45000],
    'Department': ['Marketing', 'Finance', 'Marketing', 'Finance', 'Operation'],
    'Status': ['Senior', 'Senior', 'Junior', 'Junior', 'Junior']
}
df = pd.DataFrame(data)
df_sorted = df.sort_values(by='Salary', ascending=False)
plt.bar(df_sorted['Name'], df_sorted['Salary'],color="#005f73",width=0.5)
plt.xlabel('Name of emp   loyees',fontweight ='bold')
plt.ylabel('Salary',fontweight ='bold')
plt.title('Salary track of employees',fontweight ='bold')
plt.show()
plt.xticks(rotation=0)
average_salary= df.groupby('Department')['Salary'].mean()
average_salary.plot(kind='bar', color='skyblue')

plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salary by Department')
plt.xticks(rotation=0)
plt.show()

average_salary.plot(kind='pie',)
plt.show()
```

```python
# question - 2
import pandas as pd
import matplotlib.pyplot as plt

# Organize the data into lists
data = {
    'Name': ['Rega', 'Linda', 'John', 'Vishal', 'Mohan'],
    'Department': ['Marketing', 'Finance', 'Marketing', 'Finance', 'Operation'],
    'Status': ['Senior', 'Senior', 'Junior', 'Junior', 'Junior'],
    'Salary': [30000, 40000, 100000, 75000, 45000],
    'Age': [40, 26, 21, 23, 35]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by the "Salary" column
sorted_df = df.sort_values(by='Salary')

plt.plot(sorted_df['Name'], sorted_df['Salary'], marker='o', color='b', linestyle='-')
plt.title('Employee Salaries')
plt.xlabel('Employee Name')
plt.ylabel('Salary ($)')

plt.grid()
plt.show()
```

```python
"""
question -4 
Write a program code to:
(1) Create a dataframe to store the sales record data shown in table above.
(2) Transform all the dates in "Date' Column Into format of type 'March 7,2023'
(3) Detect presence of any missing or invalid entries in the dataset Replace/Fill missing / invalid entries in product1 and Product2 column with Value (25)
(4) write a program code to perform perform following functions on the modified dataset
   (1) Analyze sales of though the chort 'Product 1' across one week through Pie chart"""
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dictionary to store the data
data = {
    'Date': ['07-03-2023', '08-03-2023', '09-03-2013', '10-03-2023', '11-03-2023', '12-03-2023', '13-03-2023'],
    'Product 1': [100, 70, None, 125, 65, 30, 50],
    'Product 2': [20, 78, 65, 100, None, 25, 55]
}
df = pd.DataFrame(data)
print(df)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y').dt.strftime('%B %d, %y')
df = df.fillna(25)
print(df)
plt.pie(df['Product 1'], labels=df['Date'])
plt.title('Sales of Product 1 Across One Week')
plt.show()
```
